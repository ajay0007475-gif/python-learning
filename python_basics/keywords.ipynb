# Keywords

## Keywords are reserved words in Python that have a special meaning and purpose.

## We cannot use them as variable names because Python already uses them to define the structure and logic of the program.

## Think of keywords like special command words in Python’s language.

# Why Can't We Use Them as Variable Names?

#### Because Python already understands them in a specific way.

```python
## For Example

if = 10

## This will give an error ❌
## Because if is a keyword used for conditions.
```

# Common Python Keywords

#### Some important keywords are:

#### if                                  else                               elif                                    for
#### while
#### break
#### continue
#### def
#### return
#### class
#### try
#### except
#### True
#### False
#### None
#### and
#### or
#### not
#### import
#### from

These words already have predefined meanings in Python.


```python
## Example 1: Using a Keyword Properly

if 5 > 3:
    print("5 is greater than 3")

## Here:

## if is a keyword used for conditional statements.

## Python understands that we are checking a condition.    
```

```python
## Example 2: Function Keyword

def greet():
    return "Hello"

print(greet())

## Here:

## def is used to define a function.

## return is used to send a value back.

## Both are keywords.
```

```python
## Wrong Usage Example

class = "Python"

## This will cause a syntax error because class is a keyword used for defining classes.
```
