## 1. Arithmetic Operators

```python
# Addition Operator

a = 4
b = 5
print (a + b)
```

```python
# Subtraction Operator

a = 10
b = 3
print (a - b)
```

```python
# Product Operator or Multiplication Operator

a = 4
b = 5
print (a * b)
```

```python
# Division Operator

a = 20
b = 6
print (a / b)
```

```python
a = 20
b = 6
print (a // b)
```

```python
# Modulo Operator
## Modulo Operator will be helping us out with getting  the idea about the remainder that will be Left after the dividing the first value with the second value.

a = 10
b = 3
print (a % b)
```

```python
# Exponential Operator

a = 2
b = 3
print (a ** b)
```

## 2. Comparison Operators

```python
## Greater than (>)

a = 10
b = 3
print (a > b)
```

```python
## Less than (<)
a = 10
b = 3
print (a < b)
```

```python
## Greater than or equal to (>=)
a = 10
b = 3
print (a >= b)
```

```python

## Less than or equal to (<=) 
a = 10
b = 3  
print (a <= b)
```

## 3. Logical Operators


```python
## and ---> Logical Operator
a = 10
b = 3
c = 5
print(a > b and a > c)
```

```python
## or ---> Logical Operator
a = 10
b = 3
c = 5
print(a > b or a > c)
```

```python
## not---> Logical Operator

a = 10
b = 3
c = 5
print(not (a > b and a > c))
```

# Simple Calculator

```python
a = int(input("Enter the first value: "))
b = int(input("Enter the second value: "))
operation = input("Enter the operation you want to perform (+, -, *, /, //, %, **): ")

if operation == "+":
    print ("The addition of the two values is: ", a + b)
elif operation == "-":
    print ("The subtraction of the two values is: ", a - b)
elif operation == "*":
    print ("The multiplication of the two values is: ", a * b)
elif operation == "/":
    print ("The division of the two values is: ", a / b)
elif operation == "//":
    print ("The floor division of the two values is: ", a // b)
elif operation == "%":
    print ("The modulus of the two values is: ", a % b)
elif operation == "**":
    print ("The power of the two values is: ", a ** b)
else:
    print("Invalid operation")
```
